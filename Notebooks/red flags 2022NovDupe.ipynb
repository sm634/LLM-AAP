````python
helpers.print_meminfo()
### TODO: Move to python function(s) inside ./src directory !!!
import re

# REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;.-]')
REPLACE_BY_SPACE_RE = re.compile('[/(){}[]|@,;.-]')
BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')
STOPWORDS = set(stopwords.words('english'))

def sl_cleanup(text):
    """
        text: a string
        
        return: modified initial string
    """
    text = text.lower() # lowercase text
    text = REPLACE_BY_SPACE_RE.sub(' ', text) # replace REPLACE_BY_SPACE_RE symbols by space in text
    text = BAD_SYMBOLS_RE.sub('', text) # delete symbols which are in BAD_SYMBOLS_RE from text
    text = ' '.join(word for word in text.split() if word not in STOPWORDS) # delete stopwors from text
    return text

def cleanText(sentence):
    sentence = fix_text(sentence)
    sentence = sl_cleanup(sentence)
    sentence = re.sub(r"(\n+|\t+)","",sentence, 0, re.MULTILINE)
    sentence = re.sub(r"(\.|\,)(\"|\“)","\\1 \\2",sentence, 0, re.MULTILINE)
    sentence = re.sub(r"([0-9])([a-z]|[A-Z])","\\1 \\2",sentence, 0, re.MULTILINE)
    sentence = re.sub(r"([a-z]|[A-Z])([0-9])","\\1 \\2",sentence, 0, re.MULTILINE)
    sentence = re.sub(r"([a-z]|[A-Z])(\.)([0-9])","\\1\\2 \\3",sentence, 0, re.MULTILINE)
    sentence = re.sub(r"([0-9])(\.)([a-z]|[A-Z])","\\1\\2 \\3",sentence, 0, re.MULTILINE)
    
    return sentence

helpers.print_meminfo()
````

## Identification matrix model

Model to identify typologies and red flags using TAH's classified news data.

We decided on a matrix identification approach using three levels of identification:

Primary Flags - these are flags with a strong connection based on TAH classified trafficking types.
Secondary Flags - these are flags with a weaker connection based on TAH classified trafficking types.
Keyword Flags - these are flags based on keywords within the text of the article.

Flags have been identified from various available collections based on expert research.

The different methods are combined to find the top three flags for each article and each unique combination of flags is combined to make a typology.

````python
# Find primary, secondary and keyword flags in article collections

articles_flags = []

for article in articles:
    # print(article)
    
    # add primary financial flags
    fin_flags_primary = []
    
    try:
        for trafftype in article["trfk_type"]["keys"]:
            trafftype = trafftype.strip().replace("  ", " ")
            if trafftype in financial_flags_primary.index and trafftype != "":
                fin_flags_primary.extend(financial_flags_primary.loc[[trafftype]].stack().reset_index(level=1).groupby(level=0, sort=False)['level_1'].apply(list)[trafftype])
            
    except KeyError:
        pass
    
    article["fin_flags_primary"] = fin_flags_primary
    
    # add secondary finanical flags
    fin_flags_secondary = []
    
    try:
        for trafftype in article["trfk_type"]["keys"]:
            trafftype = trafftype.strip().replace("  ", " ")
            if trafftype in financial_flags_secondary.index and trafftype != "":
                fin_flags_secondary.extend(financial_flags_secondary.loc[[trafftype]].stack().reset_index(level=1).groupby(level=0, sort=False)['level_1'].apply(list)[trafftype])
            
    except KeyError:
        pass
    
    article["fin_flags_secondary"] = fin_flags_secondary
    
    # add keyword flags
    article["fin_flags_keywords"] = df_train.loc[df_train['_id'] == article['_id']]['finance'].tolist()
    article["fin_flags_keywords"] = article["fin_flags_keywords"][0]
    
    articles_flags.append(article)
        
````

````python
# create dictionaries of flags and categories

# flags as key, categories as values
flag_categories = dict()

for flag in keyword_flags.columns:
    flag_categories[flag] = list(keyword_flags[keyword_flags[flag].notnull()][flag].index)

# categories as key, flags as values
categories_flag = dict()

for category in keyword_flags.index:
    categories_flag[category] = list(keyword_flags.columns[list(keyword_flags.loc[category].notnull())])


````

````python
def primary_flags(primary, types):
    """
    function to apply scores to primary flags
    takes in list of primary flags and trafficking types and returns flags with scores
    returns list of dicts
    """
    # empty list to hold results
    scored_flag_list = []
    
    # for each flag look up which trafficking type it relates to and find the confidence score - this is TAH classifier confidence
    for flag in primary:
        scored_flag = {}
        trafficking_types = financial_flags_primary.index[financial_flags_primary[flag] == 'x'].to_list()
        for trafficking_type in trafficking_types:
            try:
                dict_index = types['keys'].index(trafficking_type)
                score = types['scores'][dict_index]
            except ValueError:
                pass
        scored_flag['name'] = flag
        scored_flag['relevance'] = float(score)*0.9
        
        if len(scored_flag) > 0:
            scored_flag_list.append(scored_flag)
    
    return scored_flag_list

````

````python
def secondary_flags(secondary, types):
    """
    function to apply scores to secondary flags
    takes in list of secondary flags and trafficking types and returns flags with scores
    returns list of dicts
    behaviour is similar to primary_flags() but returns a lower score
    """
    # empty list to hold results
    scored_flag_list = []
    
    # for each flag look up which trafficking type it relates to and find the confidence score - this is TAH classifier confidence
    for flag in secondary:
        scored_flag = {}
        trafficking_types = financial_flags_secondary.index[financial_flags_secondary[flag] == 'x'].to_list()
        for trafficking_type in trafficking_types:
            try:
                dict_index = types['keys'].index(trafficking_type)
                score = types['scores'][dict_index]
            except ValueError:
                pass
        scored_flag['name'] = flag
        scored_flag['relevance'] = float(score)*0.66 #confidence score is lowered because flags are secondary
        
        if len(scored_flag) > 0:
            scored_flag_list.append(scored_flag)
    
    return scored_flag_list

````

````python
def keyword_flags(keywords, primary):
    """
    function for matching primary flags and keyword flags. This function applies a high score to any primary flag that is associated
    with a category of flag that is also indicated by the keyword analysis.
    takes in lists of primary and keyword flags
    returns list of dicts
    """
    scored_flag_list = []

    # identify joint flags
    joint_flags = []
    for word in keywords:
        joint_flags.extend(list(set(article['fin_flags_primary']) & set(categories_flag[word])))
    
    scored_flag = {}
    
    for flag in joint_flags:
        scored_flag['name'] = flag
        scored_flag['relevance'] = float(1) # allocae a score of 1
    
    if len(scored_flag) > 0:
        scored_flag_list.append(scored_flag)
        
    return scored_flag_list

````

````python
def top_three(scored_flags):
    """
    Function for identifying the three highest scoring flags
    Takes in list of flags and returns the top three
    Room for improvement - currently no way of distinguishing between multiple flags with same score
    """

    # reorder by relevance
    sorted_flags = sorted(scored_flags, key=lambda d: d['relevance'], reverse=True)
    
    # initialise empty results list
    best_three_flags = []
    
    # take top three, excluding any doubles
    for flag in sorted_flags:
        if len(best_three_flags) < 3:
            if flag['name'] in set((i['name'] for i in best_three_flags)):
                pass
            else:
                best_three_flags.append(flag)

    return best_three_flags
````

````python
def identify_typology(three_flags):
    """
    function for identifying typologies. 
    Takes in three top flags and the typology library.
    Returns typology name
    """

    # reads in typology library. read and write currently done within the function which is labour intensive but ensures library
    # is always up to date.
    # try:
       #with helpers.get_mongo_client(TLS_ENABLED) as client:
            # https://pymongo.readthedocs.io/en/stable/api/pymongo/collection.html#pymongo.collection.Collection.find
            # db = client.infinitech3
            # typologies = db.typologies.find()
            #typologies = list(typologies)
    # except Exception as e:
        # print('EXCEPTION', e)  
    try:
        dbname = get_database()
        typologies = list(dbname["ip3_typologies"].find())
    except Exceptionas as e:
        print('Exception',e)
        
    # create list of flags and checks if it's already in the typology library
    flagslist = sorted([d['name'] for d in three_flags])
    if flagslist != []:
        if flagslist not in [d['flags'] for d in typologies]:
            # if not creates a new name and writes it to the library. Name is mixture of random string of numbers and first word of each flag
            try:
                typology_name = "typ-" + str(randint(10000, 99999)) + "-" + flagslist[0].split()[0] + "-" + flagslist[1].split()[0] + "-" + flagslist[2].split()[0]
            except IndexError:
                try:
                    typology_name = "typ-" + str(randint(10000, 99999)) + "-" + flagslist[0].split()[0] + "-" + flagslist[1].split()[0]
                except IndexError:
                    typology_name = "typ-" + str(randint(10000, 99999)) + "-" + flagslist[0].split()[0]
            new_typology = {'name': typology_name,
                               'flags': flagslist}
            try:
                with helpers.get_mongo_client(TLS_ENABLED) as client:
                    data = new_typology
                    result = client.infinitech3.typologies.insert_one(data)
                    print(result)
            except Exception as e:
                print('EXCEPTION', e)
                
        # if the typology already exists, the name is extracted
        else:
            for typology in typologies:
                if typology['flags'] == flagslist:
                    typology_name = typology['name']
        return typology_name
````

````python
#All brought together here. For each article primary secondary and keyword functions executed then

# Export new results
results = []

for article in articles_flags:
    
    # find all flags from primary secondary and keyword method
    all_flags = []
    
    try:
        all_flags.extend(primary_flags(article['fin_flags_primary'], article['trfk_type']))
    except KeyError:
        pass

    try:
        all_flags.extend(secondary_flags(article['fin_flags_secondary'], article['trfk_type']))
    except KeyError:
        pass
    
    try:
        all_flags.extend(keyword_flags(article['fin_flags_keywords'], article['fin_flags_primary']))
    except KeyError:
        pass
    
    # identify top 3
    
    three_flags = top_three(all_flags)
    
    # identify typologies
    
    typology = identify_typology(three_flags)
    
    # initialise empty output dict and populate
    output = {}
    
    output['id'] = article['_id']
    
    try:
        output['frf_people'] = article['nlu']['Person']
    except KeyError:
        output['frf_people'] = []
    try:
        output['frf_company'] = article['nlu']['Company']
    except KeyError:
        output['frf_company'] = []
    try:
        output['frf_flags'] = three_flags
    except KeyError:
        output['frf_flags'] = []
    output['frf_typologies'] = typology
    output['timestamp'] = datetime.datetime.now()
    output['frf_crawl_date'] = article['crawl_date']
    

   

    results.append(output)

````
